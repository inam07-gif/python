## Stress Level Prediction – Model Analysis & Findings

###  Objective
The goal was to predict Stress_Level using smartphone usage and lifestyle features.

---

##  Correlation Analysis

The Pearson correlation between Stress_Level and all independent variables was examined.

Key observation:

- Stress_Level and Work_Productivity_Score correlation ≈ **0.0097**
- All other features also showed correlations close to **0**

 Interpretation:
- No strong **linear relationship** exists between stress and the available features.

---

##  Linear Regression Results

A Linear Regression model was trained using all features.

**Result:**
- R² ≈ **0.00**

 Interpretation:
- The model performs no better than predicting the mean stress level.
- This confirms the absence of a linear relationship.

---

## Random Forest Results (With Productivity)

A Random Forest Regressor was trained using all features.

**Result:**
- R² ≈ **0.85**

 Interpretation:
- The model achieved very high predictive performance.
- This suggests the presence of a **strong non-linear relationship** in the data.

---

##  Random Forest Results (Without Productivity)

To test feature influence, Work_Productivity_Score was removed and the model was retrained.

**Result:**
- R² ≈ **0.00**

 Interpretation:
- Predictive power collapsed after removing productivity.
- This indicates that **Work_Productivity_Score is the primary driver of Stress_Level prediction.**
- Other lifestyle and phone usage features have negligible predictive impact.

---

## Final Conclusion

- Stress_Level appears to depend almost entirely on Work_Productivity_Score.
- The relationship is **non-linear**, which explains why:
  - Linear Regression failed.
  - Random Forest performed well.
- Other smartphone usage features do not meaningfully contribute to stress prediction in this dataset.

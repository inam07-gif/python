```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.preprocessing import OneHotEncoder, PowerTransformer
from sklearn.impute import SimpleImputer

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import ElasticNet
from sklearn.preprocessing import StandardScaler
```

```python
x_train , x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
pipe_elastic=Pipeline([
    ('Scaler',StandardScaler()),
    ('poly',PolynomialFeatures()),
    ('model',ElasticNet())
],

)
```

```python
paran_grid_elastic=[{
    'poly__degree':[1,2,3,4],
    'model__alpha':[0.1,0.6,1,2,3],
    'model__l1_ratio':[0.1,0.3,0.5,0.7]}]
```

```python
grid_elastic=GridSearchCV(
    estimator=pipe_elastic,
    param_grid=paran_grid_elastic,
    cv=5,
    scoring='r2')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
best_estimator_elastic=grid_elastic.fit(x_train,y_train)
```

```python
best_estimator_elastic.best_score_
```

```python
best_estimator_elastic.best_estimator_
```

```python
y_pred_elastic=best_estimator_elastic.predict(x_test)
```

```python
mae_el = mean_absolute_error(y_test, y_pred_elastic)
rmse_el = np.sqrt(mean_squared_error(y_test, y_pred_elastic))
mse_el = mean_squared_error(y_test, y_pred_elastic)
r2_el=r2_score(y_test,y_pred_elastic)
```

```python
data_el={'Model' :['MAE','MSE','RMSE','R2_SCORE'],
         'Values':[mae_el,mse_el,rmse_el,r2_el]}
dataFrame_el=pd.DataFrame(data_el)
```

```python
dataFrame_el
```

#The ElasticNet regression model:
**A**chieved an R² score of about 0.96, meaning it explains approximately 96% of the variance in the target variable. This indicates that the model captures most of the underlying patterns in the data.

The MAE (0.13) shows that the model’s predictions deviate from the actual values by about 0.13 units on average, indicating relatively small prediction errors. Meanwhile, the RMSE (~0.17) is slightly higher because it penalizes larger errors more heavily.

Overall, the ElasticNet model demonstrates strong predictive performance, suggesting that the combination of L1 and L2 regularization effectively balances bias and variance.

**NOW SVR**

```python
svr_pipeline=Pipeline([
    ('scaler',StandardScaler()),
    ('model_svr',SVR())
])
```

```python
param_grid_svr = [

{
    'model_svr__kernel': ['linear'],
    'model_svr__C': [0.1, 1, 3]
},

{
    'model_svr__kernel': ['rbf'],
    'model_svr__C': [0.1, 1, 3],
    'model_svr__gamma': ['scale', 0.1, 1]
},

{
    'model_svr__kernel': ['poly'],
    'model_svr__C': [0.1, 1, 3],
    'model_svr__degree': [1,2],
    'model_svr__gamma': ['scale', 0.1]
}

]
```

```python
grid_svr=GridSearchCV(
    estimator=svr_pipeline,
    param_grid=param_grid_svr,
    scoring='r2',
    cv=3,
    verbose=2,
    n_jobs=-1

)
```

```python
estimator_svr=grid_svr.fit(x_train,y_train)
```

```python
estimator_svr.best_estimator_
```

```python
estimator_svr.best_score_
```

```python
y_pred_svr=estimator_svr.predict(x_test)
```

```python
mae_svr=mean_absolute_error(y_test,y_pred_svr)
rmse_svr=np.sqrt(mean_squared_error(y_test,y_pred_svr))
mse_svr=mean_squared_error(y_test,y_pred_svr)
r2_svr=r2_score(y_test,y_pred_svr)
```

```python
data_svr={'Model Metrics':['MAE','MSE','RMSE','R2_SCORE'],
         'Values':[mae_svr,mse_svr,rmse_svr,r2_svr]}
dataFrame_svr=pd.DataFrame(data_svr)
```

```python
dataFrame_svr
```

**The SVR (Support Vector Regression) model**
achieved an R² score of approximately 0.96, meaning it explains about 95.9% of the variance in the target variable. This indicates that the model successfully captures most of the relationships between the input features and the target.

The MAE (0.145) shows that the model’s predictions differ from the actual values by about 0.15 units on average, indicating relatively small prediction errors. Meanwhile, the RMSE (0.182) is slightly higher because it penalizes larger prediction errors more strongly.

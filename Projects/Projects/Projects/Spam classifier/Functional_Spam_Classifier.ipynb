```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
def transformer (data):
    data = data.lower()
    data =re.sub(r'[^a-z0-9\s]', '', data)
    data=word_tokenize(data)

    stop_words = set(stopwords.words('english'))
    data=[word for word in data if word not in stop_words]
    stemming=PorterStemmer()
    data=[stemming.stem(word) for word in data]
    return ' '.join(data)
```

```python
df.loc[:,'Clean']=df['Message'].apply(transformer)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfid=TfidfVectorizer(ngram_range=(1,2))
```

```python
x=df['Clean']
y=df['Target']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
x_train_tfid = tfid.fit_transform(x_train)
x_test_tfid  = tfid.transform(x_test)
```

```python
x_train_tfid
```

```python
x_test_tfid
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB(alpha=0.08)
model.fit(x_train_tfid, y_train)
```

```python
y_pred = model.predict(x_test_tfid)
```

```python
y_pred
```

```python
user_input = input("Enter a message: ")

clean_input = transformer(user_input)

input_tfidf = tfid.transform([clean_input])

prediction = model.predict(input_tfidf)


predictions=model.predict_proba(input_tfidf)[0][1]
if predictions>=0.5:
    print("🚨 Spam message")
else:
    print("✅ Not spam (Ham)")
```
